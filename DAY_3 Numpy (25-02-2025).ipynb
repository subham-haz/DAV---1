# Shallow Copy and Deep Copy

```python
import numpy as np

a = np.array([1,2,3])
b = a        # shallow copy
c = a.view() # shallow copy
d = a.copy() # deep copy
e = a[0:3]   # shallow copy

print(id(a), id(b))  # same object
print(id(c))         # different object
print(id(d))         # different object
print(id(e))         # different object
```

```python
d[2] = 30
d[1] = 10
e[0] = 99
a,b,c,d,e
```

```python
np.shares_memory(a,b)
```

```python
np.shares_memory(a,e)
```

```python
h = a[a%1 == 0]
```

```python
np.shares_memory(a,h)
```

# Stacking

```python
a = np.array([1,2,3])
b = np.array([4,5,6])

result = np.vstack((a, b))
result
```

```python
np.hstack((a,b))
```

```python
np.concatenate((a,b), axis=0)
```

```python
a = np.array([[1,2],[3,4]])
b = np.array([[5,6],[7,8]])

np.hstack((a,b))
```
