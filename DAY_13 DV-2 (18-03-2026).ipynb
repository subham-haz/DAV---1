```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('final_vg.csv')
data.head(2)
```

```python
data.shape
```

```python
data['Name'].value_counts()
```

```python
ih = data[data['Name'] == 'Ice Hockey']
bb = data[data['Name'] == 'Baseball']
```

```python
sns.countplot(data=data, x="Genre")
plt.xticks(rotation=45)
```

```python
sns.lineplot(data=ih, x='Year', y='NA_Sales')
# plt.xlim(2000,2010)
# plt.ylim(0.4,0.5)
```

```python
sns.lineplot(data=bb, x='Year', y='NA_Sales')
```

```python
sns.lineplot(data=ih, x='Year', y='NA_Sales', color='red', label='ice hockey')
sns.lineplot(data=bb, x='Year', y='NA_Sales', color='green', label='baseball')
plt.legend(loc='lower left', frameon=False)
plt.show()
```

```python
data[['Rank', 'Global_Sales']].corr()
```

```python
sns.scatterplot(data=data, x='Rank', y='Global_Sales')
```

```python
top_pub = data['Publisher'].value_counts().index[0:3]
top_gen = data['Genre'].value_counts().index[0:3]
top_pl = data['Platform'].value_counts().index[0:3]


data_top_pub=data[data['Publisher'].isin(top_pub)]
data_top_gen=data[data['Genre'].isin(top_gen)]
data_top_pl=data[data['Platform'].isin(top_pl)]

data_top = data[(data['Publisher'].isin(top_pub))
              & (data['Genre'].isin(top_gen))
              & data['Platform'].isin(top_pl)]
```

```python
data_top.shape
```

```python
sns.countplot(data=data_top, x='Publisher', hue='Genre')
```

```python
sns.boxplot(data=data_top, y='Global_Sales')
```

```python
sns.boxplot(data=data_top, y='Global_Sales', x='Publisher')
```

As per above graph EA is the best performer globally.

```python
sns.boxplot(data=data_top, y='Global_Sales', x='Publisher', hue='Genre')
```

```python
sns.histplot(data=data_top, x='Global_Sales', kde=True)
```

```python
sns.kdeplot(data=data_top, x='Global_Sales', hue='Publisher')
```

Density of EA sales is on right side (Performing Good)

Density of namco is on left side

```python
sns.histplot(data=data_top, x='Global_Sales', kde=True, hue='Publisher')
```

```python
sns.barplot(data=data_top, x='Publisher', y='Global_Sales', estimator=np.sum)
plt.show()
```

```python
sns.barplot(data=data_top, x='Publisher', y='Global_Sales', estimator=np.median)
```

```python
sns.barplot(data=data_top, x='Publisher', y='Global_Sales', estimator=np.mean)
```

```python
sns.barplot(data=data_top, x='Publisher', y='Global_Sales', hue='Genre')
```
