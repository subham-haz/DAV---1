```python
import pandas as pd
def add_and_remove(df, data, out):
    '''
    Input:
    df -> original dataframe
    data -> list of lists (rows to append)
    out -> index to remove
    
    Output:
    updated dataframe
    '''
    
    # Convert data to DataFrame with same columns
    new_rows = pd.DataFrame(data, columns=df.columns)
    
    # Append rows
    df = pd.concat([df, new_rows], ignore_index=True)
    
    # Remove row at index 'out'
    df = df.drop(index=out)
    
    # Reset index (important in most test cases)
    # df = df.reset_index(drop=True)
    
    return df
```
