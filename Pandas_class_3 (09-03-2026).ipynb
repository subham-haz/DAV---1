# Group based filtering

```python
# Get details of movie by high budget directors only
data_budget = data.groupby('director_name')['budget'].max().reset_index()
names = data_budget[data_budget['budget'] >= 100000000]
names
```

```python
data[data['director_name'].isin(names['director_name'])]
```

```python
avg_rev_dir = data.groupby('director_name')['revenue'].mean().reset_index()
avg_rev_dir.head(3)
```

```python
avg_rev_dir[avg_rev_dir['director_name'] == 'Adam McKay']['revenue']
```

```python
def get_risky_movies(row):
  avg_rev = avg_rev_dir[avg_rev_dir['director_name'] == row['director_name']]['revenue'].iloc[0]
  return 1 if row['budget'] > avg_rev else 0
```

```python
data['risky'] = data.apply(get_risky_movies, axis=1)
```

```python
data['risky'].value_counts()
```
